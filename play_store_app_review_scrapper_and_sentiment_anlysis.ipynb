Sentiment Analysis with VADER
**Sentiment Analysis with VADER
**
VADER (Valence Aware Dictionary and sEntiment Reasoner) is a pre-trained sentiment analysis model for text sentiment analysis. The model relies on the vader_lexicon dataset we downloaded earlier. When we give a string of words, VADER returns a dictionary containing the four scores:

    neg — negative
    neu — neutral
    pos — positive
    compound — (normalization of the three scores above)

```python
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()
res = [*new1['content'].apply(sid.polarity_scores)]
print(res[:3])
# output
#[{'compound': 0.6369, 'neg': 0.0, 'neu': 0.865, 'pos': 0.135},
 #ompound': -0.9371, 'neg': 0.627, 'neu': 0.274, 'pos': 0.099},
#{'compound': -0.8205, 'neg': 0.134, 'neu': 0.826, 'pos': 0.041}]
```

Above you can see VADER in action. Applying polarity_scores method to each row of our dataframe news, we get a list of dictionaries containing the four scores. We can create a data frame using from_records and concatenate to our original data frame using inner join.


```python
sentiment_df = pd.DataFrame.from_records(res)
new2 = pd.concat([new1, sentiment_df], axis=1, join="inner")
new2.head()
```

Now that we have the scores matching each comment, the next step is to label each comment as to positive, neutral or negative.

Choosing the threshold

Although the VADER Github readme informs that the threshold compound score to be 0.05, I have decided to use 0.2 as my threshold after some experiment.

```python
THRESHOLD = 0.2

conditions = [
    (new2['compound'] <= -THRESHOLD),
    (new2['compound'] > -THRESHOLD) & (new2['compound'] < THRESHOLD),
    (new2['compound'] >= THRESHOLD)
]

values = ["neg", "neu", "pos"]
new2['label'] = np.select(conditions, values)

new2.head()
```

```python
stop_words = stopwords.words('english')
print(stop_words[:10])

#['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're"]
```

```python
def custom_tokenize(text):
    # change all comments to lower case
    text = text.lower()
    
    # remove numbers
    re.sub("[0-9]+", "", text)
    
    # split on words only
    tk = nltk.tokenize.RegexpTokenizer(r'\w+')
    tokens = tk.tokenize(text)
    
    # remove stop words
    words = [w for w in tokens if not w in stop_words]
    
    return words
```

Notice that in our custom_tokenize function, I converted all comments into lowercase using Python’s .lower method.

Also, I have removed all numbers in the text using sub method from the re package we have imported earlier

```python
from itertools import chain
```

```python
def token_to_words(df, label):
    contents = df[new2['label'] == label].content
    tokens = contents.apply(custom_tokenize)
    words = list(chain.from_iterable(tokens))
    return words

pos_words = token_to_words(new2, 'pos')
neg_words = token_to_words(new2, 'neg')
```

Using Panda’s apply function, we can apply our custom function to each comment in our data frame.

Since tokens is a nested list, we use chain method from the itertools library to merge nested list into a single list.

At the end, we have two lists containing words of comments that were labelled positive or negative.

Usually, a more effective way of visualizing word tokens is by using a word cloud. A word cloud is a collection of words depicted in different sizes, where the size of words correspond to their count.

Creating a WordCloud

Word Cloud or Tag Clouds is a visualization technique for texts that are natively used for visualizing the tags or keywords from the websites. These keywords typically are single words that depict the context of the webpage the word cloud is being made from. These words are clustered together to form a Word Cloud.
